```python
import json

with open("../test_yelp_classification/checkpoint-4/trainer_state.json", "r") as f:
    trainer_state = json.load(f)
```

```python
log_history = trainer_state["log_history"]
for entry in log_history:
    print(entry)
```

```python
import matplotlib.pyplot as plt

# Separate logs into training and evaluation
train_loss = [(entry["step"], entry["loss"]) for entry in log_history if "loss" in entry]
eval_acc = [(entry["step"], entry["eval_accuracy"]) for entry in log_history if "eval_accuracy" in entry]

# Plot training loss
if train_loss:
    steps, losses = zip(*train_loss)
    plt.figure()
    plt.plot(steps, losses, label="Training Loss")
    plt.xlabel("Step")
    plt.ylabel("Loss")
    plt.title("Training Loss")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot evaluation accuracy
if eval_acc:
    steps, accuracies = zip(*eval_acc)
    plt.figure()
    plt.plot(steps, accuracies, label="Evaluation Accuracy")
    plt.xlabel("Step")
    plt.ylabel("Accuracy")
    plt.title("Evaluation Accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()
```
